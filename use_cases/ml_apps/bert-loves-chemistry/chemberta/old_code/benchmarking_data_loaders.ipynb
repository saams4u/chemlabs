```python
from nlp import load_dataset
import time
```

```python
csv_dataset = load_dataset("csv", data_files="../data/pubchem_descriptors/shard_00.csv")
```

```python
text_dataset = load_dataset("text", data_files="../data/pubchem_descriptors/shard_00.csv")
```

```python
csv_train = csv_dataset["train"]
text_train = text_dataset["train"]
```

```python
csv_times = []
text_times = []
for i in range(1000):
    s = time.time()
    csv_train[i]
    csv_times.append(time.time()-s)

    s = time.time()
    text_train[i]
    text_times.append(time.time()-s)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.distplot(csv_times, label="csv times")
sns.distplot(text_times, label="text times")
plt.legend()
plt.show()
```
